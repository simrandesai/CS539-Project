```python
import os
import torch.nn as nn
import torch
import pandas
import numpy
import matplotlib.pyplot as plt
import seaborn
import pandas as pd
from tqdm.notebook import tqdm
import random
from sklearn.model_selection import train_test_split
import torchvision.transforms.functional as TF
%matplotlib inline
```

```python
dataset_folder = '/mnt/data/BraDD-S1TS_zenodo'
meta_csv_path = os.path.join(dataset_folder, 'meta.csv')
```

```python
if os.path.exists(meta_csv_path):
    meta_df = pd.read_csv(meta_csv_path, index_col=0)
    print("Metadata loaded:")
    display(meta_df.head())
    print(f"Total samples: {meta_df.shape[0]}")
else:
    print("Meta CSV not found. Skipping metadata load.")
```

```python
def read_sample(idx):
    file = meta_df['file'].iloc[idx]
    return torch.load(os.path.join(dataset_folder, 'Samples', file), weights_only=False)
```

```python
change_indices = []
nochange_indices = []

for i in range(len(meta_df)):
    lbl = read_sample(i)['label'][1]
    if lbl.sum() > 0:
        change_indices.append(i)
    else:
        nochange_indices.append(i)

print(f"Change samples: {len(change_indices)} | No-change samples: {len(nochange_indices)}")
```

```python
# step 1 - need to select a subset of samples and then split into train, val, and test
random.seed(42)

n_change = min(1000, len(change_indices))
n_nochange = min(1000, len(nochange_indices))

subset_change = random.sample(change_indices, n_change)
subset_nochange = random.sample(nochange_indices, n_nochange)

subset_indices = subset_change + subset_nochange
subset_labels = [1] * len(subset_change) + [0] * len(subset_nochange)  # 1 = change, 0 = no-change

# Shuffle while keeping labels aligned
combined = list(zip(subset_indices, subset_labels))
random.shuffle(combined)
subset_indices, subset_labels = zip(*combined)

print(f"Total subset: {len(subset_indices)} samples")


# Split 80% train+val, 20% test (stratified)
train_val_idx, test_idx, train_val_y, test_y = train_test_split(
    subset_indices, subset_labels, test_size=0.2, random_state=42, stratify=subset_labels
)

# Split 10% of train_val into validation (stratified again)
train_idx, val_idx, train_y, val_y = train_test_split(
    train_val_idx, train_val_y, test_size=0.1, random_state=42, stratify=train_val_y
)

print(f"Train: {len(train_idx)} ; Val: {len(val_idx)} ; Test: {len(test_idx)}")
# should be 2000 randomized samples, but need to figure out a way to fairly determine the split between change 
# and no change samples (right now 50% each)
```

```python
#double checking the class distributions look right (should be 50/50 split????) <-- might need to change
def label_ratio(labels):
    total = len(labels)
    return f"{sum(labels)} change ({sum(labels)/total*100:.1f}%) | {total - sum(labels)} no-change ({(total-sum(labels))/total*100:.1f}%)"

print("Train:", label_ratio(train_y))
print("Val:", label_ratio(val_y))
print("Test:", label_ratio(test_y))
```

```python
#step 2 - data normalization for the 2 polarization channels --> VV & VH. (measured in dB which usually range from -50 to 0)
MIN_VALUE = -50.0
MAX_VALUE = 0


def normalize_sample(sample):
    img = sample['image']  # (T, 2, H, W)
    lbl = sample['label']  # (2, H, W)
    
    img_norm = torch.clamp((img - MIN_VALUE) / (MAX_VALUE - MIN_VALUE), 0.0, 1.0)
    
    return {
        'image': img_norm,
        'label': lbl
    }
```

```python
#step 3 - data augmentation! (need to decide which ones to use here!)
def augment_sample(sample, p_flip = 0.7, p_rotate = 0.5):
    img, lbl = sample["image"], sample["label"]

    T, C, H, W = img.shape

    #horizontal flip
    if random.random() < p_flip:
        img = torch.flip(img, dims=[3])
        lbl = torch.flip(lbl, dims=[2])

    #vertical flop
    if random.random() < p_flip:
        img = torch.flip(img, dims=[2])
        lbl = torch.flip(lbl, dims=[1])

    #rotation
    if random.random() < p_rotate:
        k = random.choice([1,2,3])  # multiples of 90
        img = torch.rot90(img, k, dims=[2,3])
        lbl = torch.rot90(lbl, k, dims=[1,2])

    return {'image': img, 'label': lbl}
```

```python
# put it all tgt and into dataloader
import torch
import random

MIN_VALUE = -50.0
MAX_VALUE = 0.0
FIXED_TIMESTEPS = 15  # new fixed sequence length

class BRADS1TDataset(Dataset):
    def __init__(self, indices, augment=True, p_flip=0.5, p_rotate=0.5):
        self.indices = indices
        self.augment = augment
        self.p_flip = p_flip
        self.p_rotate = p_rotate

    def __len__(self):
        return len(self.indices)

    def __getitem__(self, idx):
        sample_idx = self.indices[idx]
        sample = read_sample(sample_idx)

        # ----------------------
        # 1. Normalize
        # ----------------------
        img = torch.clamp((sample['image'] - MIN_VALUE) / (MAX_VALUE - MIN_VALUE), 0.0, 1.0)
        lbl = sample['label'][1].float()

        # ----------------------
        # 2. Truncate to fixed timesteps
        # ----------------------
        if img.shape[0] > FIXED_TIMESTEPS:
            img = img[:FIXED_TIMESTEPS]
        elif img.shape[0] < FIXED_TIMESTEPS:
            # pad shorter sequences with zeros
            T, C, H, W = img.shape
            pad = torch.zeros((FIXED_TIMESTEPS - T, C, H, W), dtype=img.dtype)
            img = torch.cat([img, pad], dim=0)

        # ----------------------
        # 3. Optional augmentation
        # ----------------------
        if self.augment:
            img, lbl = self.apply_augmentation(img, lbl)

        return img, lbl

    def apply_augmentation(self, img, lbl):
        T, C, H, W = img.shape

        # Horizontal flip
        if random.random() < self.p_flip:
            img = torch.flip(img, dims=[3])
            lbl = torch.flip(lbl, dims=[1])
        # Vertical flip
        if random.random() < self.p_flip:
            img = torch.flip(img, dims=[2])
            lbl = torch.flip(lbl, dims=[0])
        # Rotation 90, 180, 270
        if random.random() < self.p_rotate:
            k = random.choice([1,2,3])
            img = torch.rot90(img, k, dims=[2,3])
            lbl = torch.rot90(lbl, k, dims=[0,1])

        return img, lbl
```

```python
# Training dataset with augmentation
train_dataset = BRADS1TDataset(train_idx, augment=True)

# Validation / Test datasets without augmentation
val_dataset = BRADS1TDataset(val_idx, augment=False)
test_dataset = BRADS1TDataset(test_idx, augment=False)

print("Train samples:", len(train_dataset))
print("Validation samples:", len(val_dataset))
print("Test samples:", len(test_dataset))
```

```python
from torch.utils.data import DataLoader

batch_size = 4

train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True, num_workers=2)
val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False, num_workers=2)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=False, num_workers=2)

# Check one batch
images, labels = next(iter(train_loader))
print("Images shape:", images.shape)  # (B, T, 2, H, W)
print("Labels shape:", labels.shape)  # (B, H, W)
```
